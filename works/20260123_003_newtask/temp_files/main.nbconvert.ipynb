# UDP 生成自由能计算

计算UDP在298K，pH=10，pMg=6，I=0.3M条件下的生成自由能

```python
# 导入必要的库
from equilibrator_api import ComponentContribution, Q_
import numpy as np

# 初始化ComponentContribution对象
cc = ComponentContribution()
```

```python
# 定义计算条件
temperature = 298  # K
pH = 10
pMg = 6
ionic_strength = 0.3  # M

# 获取UDP化合物对象
udp = cc.get_compound('UDP')

# 计算在指定条件下的生成自由能
# 首先获取标准生成自由能
standard_gformation = udp.standard_dgf_prime

# 计算变换到指定条件所需的校正项
correction = udp.transform(
    p_h=Q_(pH),
    p_mg=Q_(pMg),
    ionic_strength=Q_(f'{ionic_strength}M'),
    temperature=Q_(f'{temperature}K')
)

# 计算在指定条件下的实际生成自由能
formation_energy_corrected = standard_gformation.value.m + correction.m_as('kJ/mol')

# 输出结果
print(f'UDP在{temperature}K, pH={pH}, pMg={pMg}, I={ionic_strength}M条件下的生成自由能:')
print(f'标准生成自由能: {standard_gformation.value.m:.2f} ± {standard_gformation.uncertainty.m:.2f} kJ/mol')
print(f'条件校正项: {correction.m_as("kJ/mol"):,.2f} kJ/mol')
print(f'校正后生成自由能: {formation_energy_corrected:.2f} kJ/mol')
```
